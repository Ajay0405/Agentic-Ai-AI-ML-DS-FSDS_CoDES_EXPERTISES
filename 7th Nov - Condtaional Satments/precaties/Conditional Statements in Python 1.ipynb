# If Conditional Statement

```python
age = 20
if age >= 18:
    print('Eligible to Vote')
```

```python
age = 19
if age > 18: print('Eligilble to vote')
```

# If else Conditional Statement

```python
age = 10
if age <= 12:
    print("Travel for free")
else:
    print("pay for ticket")
```

```python
# Short Hand if-else
marks = 45
res = "pass" if marks >= 40 else "Fail"
print(f"Result: {res}")
```

# elif Statement

```python
age = 25 

if age <= 12:
    print("child.")
elif age <= 19:
    print("Teenager.")
elif age <= 35:
    print("young adult.")
else:
    print("Adult")
```

# Nested if..else Conditional Statement


```python
age = 70
is_member = True

if age >= 60:
    if is_member:
        print("30% senior discount")
    else:
        print("20% senior discount")
else:
    print("Not eligible for a senior discount.")
```

# Ternary Conditional Statement

```python
# Assign a value based on a condition
age = 20
s = "Adult" if age >= 18 else "Minor"
print(s)
```

# Match-Case Statement

```python
number = 2

match number :
    case 1:
        print("One")
    case 2 | 3:
        print("Two or Three")
    case _:
        print("other number")
        
```
